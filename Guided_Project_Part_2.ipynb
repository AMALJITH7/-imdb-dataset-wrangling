```python
import pandas as pd
import numpy as np
import ast
```

# 6.Top 3 directors prefer which actors the most?

```python
data
```

```python
data=data[['crew','cast','popularity']]
data
```

```python
data['crew'].fillna('[]', inplace=True)
data['cast'].fillna('[]', inplace=True)
```

```python
df = pd.DataFrame(data)
df['crew'] = df['crew'].apply(ast.literal_eval)
df['cast'] = df['cast'].apply(ast.literal_eval)
```

```python
df['Director Name'] = df['crew'].apply(lambda x: [crew['name'] for crew in x if crew['job'] == 'Director'])
```

```python
df['Actor Name'] = df['cast'].apply(lambda x: [cast['name'] for cast in x])
```

```python
df = df[['Director Name', 'Actor Name', 'popularity']]
```

```python
df
```

```python
df['Director Name'] = df['Director Name'].apply(lambda x: ', '.join(x))
```

```python
direct_rank = df.groupby('Director Name')['popularity'].mean().reset_index()
direct_rank
```

# TOP 3 DIRECTORS ARE

```python
direct_ranks = direct_rank.sort_values(by='popularity', ascending=False)
direct_ranks.head(3)
```

```python
actors_explode=df.explode('Actor Name')
actors_explode
```

```python
actor_director=actors_explode[['Director Name','Actor Name']]
actor_director
```


# Top DirectorTim Miller prefered actor

```python
df2= pd.DataFrame(actor_director)
```

```python
Tim_Miller_df = df2[df2['Director Name'] == 'Tim Miller']
```

```python
actor_counts1= Tim_Miller_df['Actor Name'].value_counts()
```

```python
most_frequent_actor1= actor_counts1.idxmax()
most_frequent_actor_count1= actor_counts1.max()
```

```python
print(f"Director Tim Miller's prefered actor is {most_frequent_actor1}")
```

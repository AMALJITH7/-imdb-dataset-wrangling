```python
for crew_list in produceronly['crew']:
    if isinstance(crew_list, list):
        for crew_member in crew_list:
            if crew_member['job'] == 'Director':
             director_names.append(crew_member['name'])
```

```python
pd.DataFrame(director_names)
```

```python
nonenull=data4[~data4['crew'].isna()]
```

```python
nonenull
```

```python
def producers(movie):
  list1=nonenull.iloc[movie]
  details=list1.loc['crew']
  producer_name=[]
  for elem in details:
      if elem['job']=='Producer':
        producer_name.append(elem['name'])
  return producer_name
```

```python
producers(2)
```

```python
def directors(movie):
  list2=nonenull.iloc[movie]
  details2=list2.loc['crew']

  for elem in details2:
     if elem['job']=='Director':
        return elem['name']
```

```python
directors(0)
```

```python
table = pd.DataFrame(columns=['Movie name','Producers','Directors','ROI'])
```

```python
for index,row in nonenull.iterrows():

  try:
      table = table.append({'Movie name':nonenull.loc[index,'title'],'Producers':producers(index),'Directors':directors(index),'ROI':nonenull.loc[index,'ROI']},ignore_index=True)
  except:
    continue
```

# Movie with producers and director

```python
table.head()
```

# top 3 producers who have produced movies with the highest average Roi

```python
explodeproducer=table.explode('Producers')
explodeproducer
```

```python
producersroi= explodeproducer.groupby('Producers')['ROI'].mean().reset_index()
producersroi.rename(columns={'ROI':'Avg Roi'},inplace=True)
top3=producersroi.sort_values(by=['Avg Roi'],ascending=False)
top3.head(3)
```

# **Which Actor has acted the most number of movies**

```python
new_data=pd.DataFrame(actors_ind['actors'].value_counts()).reset_index()
final_data=new_data.rename(columns={'index':'actors name','actors':'number of movies'})
final_data.head(2)
```

```python
data_pro=data[['actors','genres','profit']]
actors_profit=data_pro.explode('actors','genres')
actors_profit
```

```python
robert=actors_profit.loc[actors_profit['actors'] == 'Robert De Niro']
robert
```

total profit by Robert De Niro

# Robert's movie's avg profit

```python
avg_profit_by_robert=robert.groupby('actors')['profit'].mean().reset_index()
pd.DataFrame(avg_profit_by_robert)
```

```python
robert.loc[~robert['genres'].isna(),'genres']=robert.loc[~robert['genres'].isna(),'genres']
robert['Genres']=robert['genres'].apply(lambda x:[i['name'] for i in x] if isinstance(x,list) else [])
```

```python
robert_genre=robert.drop(['genres'],axis=1)
robert_genre
```

```python
genre_count1 = {}
```

```python
for genres_list in robert_genre['Genres']:
    for genre in genres_list:
        if genre in genre_count1:
            genre_count1[genre] += 1
        else:
            genre_count1[genre] = 1
```

```python
genre_count1
```

```python
most_common_genre_by_robert = max(genre_count1, key=genre_count1.get)
most_common_genre_by_robert
```

# Which genre Robert De  Niro acts the most

```python
print(f"The most common genre for Robert De Niro is: {most_common_genre_by_robert }")
```
